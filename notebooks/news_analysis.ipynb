Task 1.D

Countries that have many articles written about them - the content of the news is about that country

Step one: since we consider title and description as the content of the news letter we should iterate through the dataset for country names and the count(how many times it appeard on newses through out the data set).

In order to do that we need list of country names to search in the news articles, and also a a dictionay to store the count 

```python
from collections import defaultdict

countries_to_search = ['United States', 'United Kingdom' 'China', 'Russia', 'Ethiopia', 'Germany', 'Kenya', 'Nigeria', 'Brazil']

# Dictionary to store country counts
country_counts = defaultdict(int)
```

now lets write the code that iterates throug the news in the data set and search the country names above and count thier frequency of apperance

```python

# Iterate over each news article
for index, row in news_data.iterrows():
    title = str(row['title']).lower()  # Convert to lowercase for matching
    description = str(row['description']).lower()
        
    for country in countries_to_search:
        if country.lower() in title or country.lower() in description:
            country_counts[country] += 1
            
# Convert country_counts to a DataFrame for easier analysis
country_counts_df = pd.DataFrame(list(country_counts.items()), columns=['Country', 'Article_Count'])

# Sort DataFrame by Article_Count in descending order
country_counts_df = country_counts_df.sort_values(by='Article_Count', ascending=False)

# Display top countries with most articles related to them
print("Top Countries with Most Articles Written About:")
print(country_counts_df.head(10))  # Display top 10 countries
```

Task 1.E                                                     Websites that reported (the news content) about Africa, US, China, EU, Russia, Ukraine, Middle East? Note that you will need to group countries together to form the African, EU, and Middle East continents/regions.

Step One: Just like the previous task(Task 1.D) this task needs  mappings countries to their respective regions (continents)

```python

# Define country-to-region mappings
country_to_region = {
    'United States': 'North America',
    'United Kingdom': 'Europe',
    'China': 'Europe',
    'Russia': 'Asia',
    'Ethiopia': 'Africa',
    'Germany': 'Europe',
    'Kenya': 'Africa',
    'Nigeria': 'Africa',
    'Brazil': 'Sout America',
    'Canada': 'North America',
    'Mexico': 'North America',
    'India': 'Asia',
    'France': 'Europe',
    'Ukraine': 'Europe',
    'South Africa': 'Africa',
    'Saudi Arabia': 'Middle East',
    'Israel': 'Middle East',
    'Iran': 'Middle East'
}
```

Step 2: Count Mentions of Regions in News Articles

```python
from collections import defaultdict
region_counts = defaultdict(int)

for index, row in news_data.iterrows():
    title = str(row['title']).lower()  # Convert to lowercase for case-insensitive matching
    description = str(row['description']).lower()

    # Check for mentions of countries and map to regions
    mentioned_regions = set()
    for country, region in country_to_region.items():
        if country.lower() in title or country.lower() in description:
            mentioned_regions.add(region)

    # Increment counts for mentioned regions
    for region in mentioned_regions:
        region_counts[region] += 1
        
# Convert region_counts to a DataFrame for easier analysis
region_counts_df = pd.DataFrame(list(region_counts.items()), columns=['Region', 'Article_Count'])

# Sort DataFrame by Article_Count in descending order
region_counts_df = region_counts_df.sort_values(by='Article_Count', ascending=False)

# Display regions with most articles related to them
print("Regions with Most Articles Written About:")
print(region_counts_df)
```
